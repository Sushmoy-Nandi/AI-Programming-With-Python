# Different type of argument


```python
def adder(a,b,c) :
  print(f"{a},{b},{c}")


adder(10,30,20)
```

```python
def adder(*args) :
  for i in args:
    print(i)


adder(10,30,20,40)
```

```python
def adder(a=10,b=30,c=10) :
  print(f"{a},{b},{c}")


adder()
```

```python
def adder(**kwargs):
  for key,value in kwargs.items():
    print(f"{key}: {value}")

adder(a = 10 , b=30 , c = 40 , n_estimator=50)
```

# generator

```python
lst = [ x for x in range(500) ] # dataset

def data_loader(chunk_size,lst):
    for i in range(0,len(lst),chunk_size):
        yield lst[i:i+chunk_size]


x = data_loader(5,lst)

print(next(x))
print(next(x))
```

# Lambda, Map and Apply function

```python
import pandas as pd


df = pd.DataFrame({'temperature (F)': [32, 212, 98.6, 68]})
df

df['temperature (C)'] = df['temperature (F)'].apply(lambda x: (x - 32) * 5/9)
df
```

```python
def adder(a,b) :
  return [a,b]

print(type(adder(10.5,20)))

#unzip

x,y = adder(10.5,20)

print(y)
```

```python
import pandas as pd


df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8]})

df

df['A'] = df['A'].apply(lambda x: x ** 0.5)


df
```

```python
import pandas as pd

df = pd.DataFrame({'total_spent': [120, 600, 450, 800]})

# Creating a 'High Value' feature on the fly
df['is_high_value'] = df['total_spent'].apply(lambda x: 1 if x > 500 else 0)

df
```
