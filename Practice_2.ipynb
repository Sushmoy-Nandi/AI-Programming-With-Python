```python
def create_file(filename, content):
    with open(filename, "w", encoding="utf-8") as file:
        file.write(content)


create_file("./sample_data/notes.txt", "Hello, this is my first file.")
```

```python
def read_lines(filename):
    with open(filename, "r") as file:
        # string_list = file.readlines()
        # total_lines = len(string_list)
        # print(total_lines)
        i = 1
        for line in file:
            print(f"{i}: {line.strip()}")
            i += 1 
    
read_lines("./sample_data/data.txt")
```

```python
def add_log(message):
    with open("./sample_data/log.txt", "a") as file:
        file.write(message + "\n")

add_log("User logged in")
add_log("File uploaded")
```

```python
def read_from_position(filename, position):
    with open(filename, "r") as file:
        file.seek(position)
        print(file.read())

read_from_position("./sample_data/hello.txt", 7)
```

```python
def safe_read(filename):
    try:
        file = open(filename,"r")
    except FileNotFoundError:
        print("Error: File not found.")
    else:
        print(file.read())
        file.close()
    finally:
        print("Operation complete.")
```

```python
safe_read("./sample_data/missing.txt")
```
