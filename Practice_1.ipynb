```python
data = [1, 2, 3, 4, 5, 6, 7, 8]
# batch_size = 3

def batch_generator(data, batch_size):
    for i in range(0, len(data), batch_size):
        yield data[i:i + batch_size]

x= batch_generator(data,batch_size=3)

print(next(x))
print(next(x))
print(next(x))
```

```python
def fun_power(base,exponent=2):
    result= base**exponent
    return result

```

```python
final_result = fun_power(5)
print(final_result)
```
